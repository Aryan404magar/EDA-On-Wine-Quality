# HEATMAP

### we use a heatmap for easier data visualation

```python
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('CORRELATION HEATMAP')
plt.show()
```

### *We can clearly see that the correlation between the quality of an alcohol and the amount of alcohol is 0.48.

### *It suggests a moderate positive linear relationship between the two variables.
### *As one variable increases, there is a tendency for the other variable to increase, but the relationship is not extremely strong.
### *The positive correlation indicates that when one variable goes up, the other variable tends to go up as well.


# To interpret a heatmap:

### * The colors represent the correlation coefficient between two variables.
### * Positive correlations are indicated by colors tending towards 1 (usually warmer colors), while negative correlations are indicated by colors tending towards -1 (usually cooler colors).
### * The annot parameter adds numerical values to each cell, indicating the correlation strength.

# Histogram

### We will also create a histogram with the help of seaborn and matplotlib

### We will create a histogram for the quality of the wine.

```python
# seaborn histogram 
sns.distplot(df['quality'], hist=True, kde=False, 
             bins=12, color = 'blue',
             hist_kws={'edgecolor':'black'})
# Add labels
plt.title('Quality count')
plt.xlabel('Quality')

plt.ylabel('Count')


```

# LINE PLOT:

```python
# line plot 
sns.lineplot(x='quality',y='alcohol', data=df )
```
